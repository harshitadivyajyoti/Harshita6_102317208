##### Q1. Ask the user to enter a value (e.g., M) for a mathematical function.  Generate x values from -10 to 10 using np.linspace().  Compute y values for: o y= M⋅x2 o y=M⋅sin(x)  Plot both functions in a single figure: o Use different colors and line styles. o Add legend, grid, and title. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
M = float(input("Enter a value for M: "))
```

```python
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y = M * x^2', color='b', linestyle='-')
plt.plot(x, y2, label='y = M * sin(x)', color='r', linestyle='--')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

##### Q.2 Create a dataset of five subjects and scores.  Convert it into a Pandas DataFrame.  Plot the scores using a Seaborn bar plot with: o Different colors for each bar. o Annotations on top of each bar. o Title, axis labels, and grid. 

```python
subjects = ['Math', 'Science', 'English', 'History', 'Computer']
scores = [85, 90, 78, 88, 92]
df = pd.DataFrame({'Subject': subjects, 'Score': scores})
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Subject', y='Score', data=df, hue='Subject', legend=False, palette='husl')
for index, value in enumerate(scores):
    plt.text(index, value + 1, str(value), ha='center')
plt.title("Subject Scores")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid()
plt.show()
```

##### Q3. Write a Python script to select your roll number as seed for NumPy and then generate a dataset of 50 values using np.random.randn().  Create a 2x2 subplot layout displaying: o Line plot showing cumulative sum. o Scatter plot with random noise.  Customize with titles, axis labels, and grids.

```python
roll_number = int(input("Enter your roll number: "))
```

```python
np.random.seed(roll_number)
```

```python
data = np.random.randn(50)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(np.cumsum(data), color='g')
axes[0, 0].set_title("Cumulative Sum")
axes[0, 0].grid()

axes[0, 1].scatter(range(50), data, color='b')
axes[0, 1].set_title("Scatter Plot")
axes[0, 1].grid()

axes[1, 0].hist(data, bins=10, color='purple', alpha=0.7)
axes[1, 0].set_title("Histogram")
axes[1, 0].grid()

axes[1, 1].boxplot(data)
axes[1, 1].set_title("Box Plot")
axes[1, 1].grid()
```

##### Q.4 Download Data-set from the below link https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv Apply ‘seaborn’ library to do the following. 1. Read Total profit of all months and show it using a line plot. 2. Read all product sales data and show it using a multiline plot. 3. Plot bar chart for all the features/attributes.

```python
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
data = pd.read_csv(url)
```

```python
data['total_profit'].plot(kind='line', marker='o', linestyle='-', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid()
plt.show()
```

```python
data.iloc[:, 1:].plot(kind='line', marker='o')
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(data.columns[1:])
plt.grid()
plt.show()
```

```python
data.iloc[:, 1:].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid()
plt.show()
```
